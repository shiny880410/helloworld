```
sh2 = gc.open_by_url('https://docs.google.com/spreadsheets/d/1_L_CG5WhF5oZUvpmJTWEbEmHHGJ1-boQ9oxyfILFJyU/edit#gid=448064943')
wks5 = sh2.worksheet("工作表2")
pi1 = wks5.col_values(1)
#print(pi1[:5])
pi2 = wks5.col_values(4)
pi3 = wks5.col_values(5)
pi4 = wks5.col_values(6)
n = 25000
pi1train = pi1[1:n]
pi2train = pi2[1:n]
pi3train = pi3[1:n]
pi4train = pi4[1:n]
pi1test = pi1[n:n+25]
pi2test = pi2[n:n+25]
pi3test = pi3[n:n+25]
pi4test = pi4[n:n+25]
```

```
dfxtrain = pd.DataFrame(
    {'pi3': pi3train,
     'pi4': pi4train,
     'pi2': pi2train
    })
dfxtrain.tail(25)
```

```
dfytrain = pd.DataFrame({'pi1':pi1train})
dfytrain.tail(25)
```

```
dfxtest = pd.DataFrame(
    {'pi3': pi3test,
     'pi4': pi4test,
     'pi2': pi2test
    })
dfxtest.tail(25)
```
