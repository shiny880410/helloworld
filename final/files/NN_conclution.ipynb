```
from google.colab import drive
import pandas as pd
drive.mount('/content/gdrive')
!cp 'gdrive/My Drive/Colab Notebooks/復興南/train data/classifier1.h5' 'classifier1.h5'
from keras.models import load_model
classifier = load_model('classifier1.h5')
```

```
occ = []
for i in range(1,101):
    occ.append(i/10)
def v(wide,redlight,turnleft):
  import math
  occ = []
  vol = 65
  length =1100
  pi2test = []
  pi3test = []
  pi4test = []
  for i in range(1,101):
    occ.append(i/10)


  for i in range(0,100):
    pi2test.append(length/wide)
    pi3test.append(math.sqrt((vol/(wide*wide))*math.sqrt(occ[i]/(redlight/length))))
    pi4test.append(math.sqrt(1/((turnleft/length)*(turnleft/length)/(wide*math.sqrt(occ[i]*redlight/length)))))
  X_test = pd.DataFrame(
      {'0': pi3test,
       '1': pi4test,
      '2': pi2test
      })
  Y_pred = classifier.predict(X_test)
  pi1 = Y_pred.tolist()
  v = []
  for i in range(100):
    v.append(pi1[i][0]*math.sqrt(redlight/length/occ[i]))
  return v
```

```
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
plt.figure(figsize=(15,15))
plt.style.use('bmh')
plt.xlim((0, 10))
#plt.ylim((30, 130))
plt.xlabel('occupation', fontsize=18)
plt.ylabel('speed', fontsize=18)
#設定粗細網格線
plt.grid(b=True, which='major', color='#666666', linestyle='-')
plt.minorticks_on()
plt.grid(b=True, which='minor', color='#999999', linestyle='-', alpha=0.2)
#畫圖
plt.scatter(occ,v(3.5,7,4),s=20, alpha=.5,c='r')
plt.scatter(occ,v(3,7,4),s=20, alpha=.5,c='b')
plt.scatter(occ,v(2,7,4),s=20, alpha=.5,c='y')
plt.scatter(occ,v(1.5,7,4),s=20, alpha=.5,c='g')
plt.xticks(range(0,10,1),range(11))
plt.legend(['lane:3.5','lane:3','lane:2','lane:1.5'],fontsize=18)
plt.title('Speed-Occ(fix vol,change lane)',fontsize=25)
plt.show()
```
